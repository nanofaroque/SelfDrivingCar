```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist_features = []   # it is a list of image matrix. Size of the image matrix is 784 (28*28). Each image is a column matrix
mnist_labels = []

mnist = input_data.read_data_sets('/Users/omarfaroque/tensorflow-master/tensorflow/examples/udacity', one_hot=True)
```

```python
n_labels = 5 # we want to work with only first five characters.
# for a our label will be like; [1,0,0,0,0]
# for b our label will be like; [0,1,0,0,0]
for mnist_feature,mnist_label in zip(*mnist.train.next_batch(10000)):
    if mnist_label[:n_labels].any(): # Since our labels are 10characters(in datatset), we will take first and check is ther any 'True' (1), 
        # it means a character which is one of the first five character. So we need to add that
        mnist_features.append(mnist_feature)
        mnist_labels.append(mnist_label[:5])
        
```
